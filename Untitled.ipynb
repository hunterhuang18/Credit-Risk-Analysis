Status : 0: 1-29 days past due 1: 30-59 days past due 2: 60-89 days overdue 3: 90-119 days overdue 4: 120-149 days overdue 5: Overdue or bad debts, write-offs for more than 150 days C: paid off that month X: No loan for the month

### Caculate the originated month and MOB for each borrower

```python
data_prep = credit_record.copy()
```

```python
data_prep = data_prep.merge(data_prep.groupby('ID')['MONTHS_BALANCE'].min(), how = 'left', left_on = 'ID', right_on = 'ID')
```

```python
data_prep.rename(columns = {'MONTHS_BALANCE_x' : 'MONTHS_BALANCE' ,'MONTHS_BALANCE_y' : 'loan_originated_month'}, inplace = True)
```

```python
data_prep = data_prep.merge(data_prep.groupby('ID')['MONTHS_BALANCE'].max(), how = 'left', left_on = 'ID', right_on = 'ID')
```

```python
data_prep.rename(columns = {'MONTHS_BALANCE_x' : 'MONTHS_BALANCE' ,'MONTHS_BALANCE_y' : 'loan_ended_month'}, inplace = True)
```

```python
data_prep['window'] = data_prep['loan_ended_month'] - data_prep['loan_originated_month']
data_prep['MOB'] = data_prep['MONTHS_BALANCE'] - data_prep['loan_originated_month']
```

```python
data_prep = data_prep[data_prep['window'] > 12]
```

```python
data_prep['status'] = np.where((data_prep['STATUS'] == '2')|(data_prep['STATUS'] == '3')|(data_prep['STATUS'] == '4')|(data_prep['STATUS'] == '5'), 1, 0)
```

```python
data_prep
```

```python
data_temp = data_prep.groupby('loan_originated_month')['ID'].nunique().reset_index()
```

```python
data_temp
```

```python
originated_month = np.repeat(range(-60, 1), 61)
MOB = np.tile(range(0, 61), 61)

vintage_data = pd.DataFrame({'originated_month' : originated_month, 'MOB' : MOB})
```

```python
vintage_data = vintage_data.merge(data_temp, how = 'left', left_on = 'originated_month', right_on = 'loan_originated_month')
vintage_data.drop('loan_originated_month', axis = 1, inplace = True)
vintage_data.rename(columns = {'ID' : 'total_accounts'}, inplace = True)
```

```python
vintage_data
```

### Calculate Delinquent accounts for each MOB and each originated_month

```python
data_prep.head()
```

```python
due_accounts = []
for j in range(-60, 1):
    ls = []
    for i in range(0, 61):
        due = list(data_prep[(data_prep['loan_originated_month'] == j) & (data_prep['MOB'] == i) & (data_prep['status'] == 1)]['ID'])
        ls.extend(due)
        due_accounts.append(len(set(ls)))    
```

```python
vintage_data['due_accounts'] = due_accounts
```

```python
vintage_data['due_rate'] = vintage_data['due_accounts']/vintage_data['total_accounts']
```
